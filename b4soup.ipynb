```python
import bs4
import pandas as pd 
import requests
```

```python
url = 'https://webtygia.com/gia-vang-sjc/ha-noi/ngay-7-02-2022.html'
```

```python
def get_page_content(url):
   page = requests.get(url,headers={"Accept-Language":"en-US"})
   return bs4.BeautifulSoup(page.text,"html.parser")
soup = get_page_content(url)
```

```python
soup
```

```python
table = []
contentTable  = soup.find('table', { "class" : "table table-striped table-hover table-bordered history2"})

import re
for a in contentTable.findAll('td', class_=['']):
  if  'Thứ hai' in a.text or 'Thứ hai' in a.text or 'Thứ ba' in a.text or 'Thứ tư' in a.text or 'Thứ năm' in a.text or 'Thứ sáu' in a.text or 'Thứ bảy' in a.text or 'Chủ nhật' in a.text:
    table.append(a.text)
    print(a.text)
```

```python
price = []

import re
contentTable  = soup.find('table', { "class" : "table table-striped table-hover table-bordered history2"}) # Use dictionary to pass key : value pair
for a in contentTable.findAll('td', class_=['text-right']):
    for text in a:
        price.append(text)
        print(a.get_text())
```
